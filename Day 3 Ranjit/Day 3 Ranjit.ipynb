## Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.array([3,13,19,31,38,45,58,65,71,88])
y = np.array([11,16,24,37,41,58,56,78,81,95])
z = np.array([81,80,79,63,54,32,29,26,18,14])
```

#### Comparing x agains y and x against z

```python
plt.title('My First Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.plot(x, y, color='red', label= '2023')
plt.plot(x, z, color='green', label= '2022')
plt.legend() ## suchi
```

#### Bar plot

```python
enrollment = [104,112,256,178,126,180]
courses = ['C','Java','Python','R','PHP','JavaScript']
```

```python
plt.ylabel('Course')  ## label exchange
plt.xlabel('Enrolled Students')
plt.grid()
plt.barh(courses, enrollment, color='brown') ### barh for horizontal 
```

## Machine Learning

### Regression

```python
# Dataset: Salary_Data.csv, mtcars.csv, Social_Network_Ads.csv, Mail_Customers.csv
# Location: https://mitu.co.in/datasets
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_Data.csv')
```

```python
df.shape
```

```python
df  ## x is input variable and y is output variable always as y depends on x
```

```python
plt.scatter(df['YearsExperience'], df['Salary']);
```

```python
x = df['YearsExperience'].values
y = df['Salary']
```

```python
x
```

```python
y
```

```python
y;
```

```python
x = x.reshape(30,1)
```

```python
x
```

#### import linear regression class

```python
from sklearn.linear_model import LinearRegression
```

#### create the object of linear regression

```python
regressor = LinearRegression()
```

#### train the algorithm

```python
regressor.fit(x,y)
```

#### predict on the unseen data

```python
regressor.predict([[15]])
```

```python
regressor.predict([[20]])
```

```python
#### accuracy of algorithm
```

```python
regressor.score(x,y)
```
